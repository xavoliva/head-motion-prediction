```python
def add_scene_cut_column(data, scene_cut_list):
    scene_cuts_column = np.zeros(len(data))
    
    index_before = 0
    scene_cut_indice = 0
    value = 0
    print(len(data))
    for row in range(len(data)):
        if data[row]['frameId'] > scene_cut_list[scene_cut_indice]:
            value = int(not bool(value))
            scene_cut_indice += 1
        print(row, value)
        scene_cuts_column[row] = value
    return rfn.append_fields(data, 'scene_cuts', scene_cuts_column)
```
