```python
import os
from os.path import isfile, join
import cv2
import matplotlib.pyplot as plt
import math
import numpy as np
import time
from scipy.io import arff
from scipy import signal, interpolate
from scipy.ndimage import gaussian_filter
import pdb
import time
from scipy.ndimage import zoom
from scipy.special import logsumexp
import tensorflow as tf
import h5py

dirpath = os.getcwd()
```

```python
def store_many_hdf5(images, directory, file_name):
    """ Stores an array of images to HDF5.
        Parameters:
        ---------------
        images       images array, (N, 32, 32, channel_size) to be stored
    """
    num_images = len(images)

    # Create a new HDF5 file
    file = h5py.File(os.path.join(directory, file_name)+".h5", "w")

    # Create a dataset in the file
    dataset = file.create_dataset(
        "images", np.shape(images), h5py.h5t.STD_U8BE, data=images
    )
    
    file.close()
```

```python
def resize_image(image, HEIGHT, WIDTH):
    resized_image = cv2.resize(image, (WIDTH,HEIGHT), interpolation=cv2.INTER_CUBIC)
    return resized_image
```

```python
start_index = 0
videos_frames_path = join(dirpath, 'datasetTUM', 'videos_frames')
videos_path = join(dirpath, 'datasetTUM', 'videos')
video_names = sorted([f for f in os.listdir(videos_path) if isfile(join(videos_path, f))])
for i in range(len(video_names)):
    print(video_names[i])
```

```python
for video_name in video_names[start_index:]:
    start_time_video = time.time()
    video_path = join(videos_path, video_name)
    print(video_name)
    
    frames = None
    vidcap = cv2.VideoCapture(video_path)
    
    success, image_frame_full = vidcap.read()
    frame_id = 0
    while success:
        #print(image_frame_full.shape)
        image_frame = resize_image(image_frame_full, int(image_frame_full.shape[0]/2), int(image_frame_full.shape[1]/2))
        #print(image_frame.shape)
        print(frame_id)
        if frames is not None:
            frames = np.vstack((frames, image_frame))
        else:
            frames = image_frame
        
        success, image_frame_full = vidcap.read()
        frame_id += 1

        
    print(frames.shape)
    store_many_hdf5(frames, videos_frames_path, video_name)
                 
    print("Time per video: " + str(time.time() - start_time_video) + " seconds,",
            str((time.time() - start_time_video)/60) + " minutes." )
```
